```python
from math import tan , pi 
def polygon_area(args,kwargs):
    length = args[0]
    sides = kwargs['sides']
    area = (sides * length**2 )/(4 *tan(pi/sides))
    return area
                             
time_it(polygon_area, 4, sides = 4, repetitons=10)
```

```python
def speed_converter(args , kwargs):
    kmph = args[0]
    dist = kwargs['dist']
    time = kwargs['time']
    
    if dist is 'km':
        d = kmph
    if dist is 'm':
        d = kmph * 1000
    if dist is 'ft':
        d = kmph * 3281
    if dist is 'yrd':
        d = kmph * 1093.61
        
    if time is 'ms':
        t = 3600002.88
    if  time is 's':
        t = 3600.00288
    if time is 'min':
        t =  60
    if time is 'hr':
        t = time
    if time is 'day':
        t = 0.0416667# ms/s/m/hr/day
        
    speed = float(d)/float(t)
    
    return speed

time_it(speed_converter, 100, dist='ft', time='s', repetitons=200)
```
